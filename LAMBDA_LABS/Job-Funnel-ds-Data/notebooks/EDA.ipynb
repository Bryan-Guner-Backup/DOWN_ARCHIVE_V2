```python
import pandas as pd
import numpy as np 
import os
from pandas_profiling import ProfileReport # requires version 2.5+
import warnings
warnings.simplefilter(action='ignore', category=Warning) 
notebook_path = os.path.abspath("EDA.ipynb")
```

```python
job_listings = os.path.join(os.path.dirname(notebook_path), "job_listings.csv")
df = pd.read_csv(job_listings, dtype={'seniority': object, 'keyphrase':object}, na_values=['--'])
df = df.rename({'Unnamed: 0':'index'}, axis=1)
```

```python
df.dtypes
```

```python
profile = ProfileReport(df)
profile
```
