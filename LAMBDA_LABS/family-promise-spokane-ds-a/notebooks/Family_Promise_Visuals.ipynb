```
import pandas as pd
import plotly.graph_objects as go
import sqlalchemy

# Makes a dataframe directly from the database
engine = sqlalchemy.create_engine('postgresql://postgres:s8!E}*ra2`*b::XQ@fam-prom-500.cbcffhx71hk4.us-east-1.rds.amazonaws.com/postgres')
df = pd.read_sql_table('members',engine)
```

```
# Create a column with just the exit years
def get_year(date):
  return date.year
df['year_of_exit'] = df['date_of_exit'].apply(get_year)
```

```
# Make a df with ideal year range
df_2020 = df[df['year_of_exit'] == 2020]
df_2020.head()
```

```
exits = ['Unkown/Other', 'Permanent Exit', 'Emergency Shelter', 'Temporary Exit', 'Transitional Housing']

# Make a pie chart of the percentage of families that exit in each manner
fig = go.Figure([go.Pie(labels=exits, values=df_2020['exit_destination'].value_counts())])
fig.update_layout(
    title={
        'text': "Exit Destination Counts in 2020",
        'y':0.9,
        'x':0.43, # was 0.5, but I changed it to be directly over the circle
        'xanchor': 'center',
        'yanchor': 'top'},
        xaxis_title="Exit Type",
        yaxis_title='Number of People')
fig.show()
```

```
# Make a dataframe comparing length of stay to how many Families have also stayed that long
df_len_stay = pd.DataFrame(df_2020['length_of_stay'].value_counts())
df_len_stay = df_len_stay.reset_index()
df_len_stay.columns= ['length_of_stay','Count']
df_len_stay
```

```

#### I'm definitely open to changes to this one. ####

# Plots a histogram and shows the average length of stay per family.
fig = go.Figure([go.Histogram(x=df_len_stay['length_of_stay'], y=df_len_stay['Count'],bingroup=4)])

# Make the average line
fig.add_shape(type='line',
                x0=df_2020['length_of_stay'].mean(),
                y0=0,
                y1=df_len_stay['Count'].max() +10,
                x1=df_2020['length_of_stay'].mean(),
                line=dict(color='Red',),
                xref='x',
                yref='y',
                name='Average'
)


fig.update_layout(
    title={
        'text': "Length of Stay Counts in 2020",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
        xaxis_title="Length of Stay",
        yaxis_title="Count (in Families)")
fig.show()
```
