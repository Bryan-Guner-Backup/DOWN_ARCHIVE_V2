# Restructure Dataset

```python
# Reset the index
rent.reset_index(level=0, inplace=True)
```

```python
# Unpivot the data on datestamp columns and its values
rent2 = rent.melt(id_vars=["index", "RegionID", "RegionName", "SizeRank"], 
        var_name="ds", 
        value_name="yhat")
rent2
```

```python
# Sort the data based on SizeRank and datestamp
rent3 = rent2.sort_values(by=['SizeRank', 'ds'])
```

# Creating predictions

```python
# Create a for loop to apply the model to all 101 cities in the dataset then concatenate the results.
from fbprophet import Prophet
for x in range(102):
    df1 = rent3.loc[rent3['index'] == x][['ds', 'yhat']]
    df2 = rent3.loc[rent3['index'] == x][['RegionID', 'RegionName', 'SizeRank']]
    df1.columns = ['ds','y']
    df1['ds'] = pd.to_datetime(df1['ds'])
    m = Prophet(interval_width=.95)
    m.fit(df1)
    future = m.make_future_dataframe(freq='MS', periods=12)
    forecast = m.predict(future)
    predictions = forecast[['ds', 'yhat']]
    predictions['yhat'] = predictions.yhat.round()
    predictions['RegionID'] = df2['RegionID'].iloc[0]
    predictions['RegionName'] = df2['RegionName'].iloc[0]
    predictions['SizeRank'] = df2['SizeRank'].iloc[0]
    predictions = predictions[['RegionID', 'RegionName', 'SizeRank', 'ds', 'yhat']]
    predictions.to_csv('predictions.csv', mode='a', index=False)
```
